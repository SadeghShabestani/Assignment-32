```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
width_balloons = np.random.normal(40, 4,500)
length_balloons = np.random.normal(10, 3,500)
weight_balloons = np.random.normal(0.5, 0.02,500)


width_melon = np.random.normal(35, 4,500)
length_melon = np.random.normal(8, 2,500)
weight_melon = np.random.normal(10, 0.5,500)
```

```python
fig = plt.figure(figsize = (15,10))
ax = plt.axes(projection ="3d")
ax.set_zlabel('weight')

ax.scatter3D(width_balloons,length_balloons , weight_balloons , color='blue' , label = 'Balloons')
ax.scatter3D(width_melon,length_melon,weight_melon , color ='green' , label = 'Melon')

plt.legend()
plt.xlabel('Width')
plt.ylabel('Height')
plt.title('Length and width of balloons and melons')
```
